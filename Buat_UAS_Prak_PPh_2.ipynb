## Data loading

### Subtask:
Load the data from "Data_Analitik_PPh_Badan_2023-2027.csv" into a pandas DataFrame.


**Reasoning**:
Import pandas and load the data into a DataFrame, then display the first 5 rows.



```python
import pandas as pd

df = pd.read_csv("Data_Analitik_PPh_Badan_2023-2027.csv")
display(df.head())
```

## Data exploration

### Subtask:
Explore the data to understand its structure, content, and potential areas for analysis and visualization. This includes checking data types, looking at summary statistics, and identifying potential issues like missing values.


**Reasoning**:
Display the shape of the DataFrame, print data types, generate descriptive statistics for numerical columns, and check for missing values to explore the data's structure and content as per the instructions.



```python
print("Shape of the DataFrame:", df.shape)
print("\nData types of each column:")
print(df.dtypes)
print("\nDescriptive statistics for numerical columns:")
display(df.describe())
print("\nMissing values per column:")
print(df.isnull().sum())
```

## Data cleaning

### Subtask:
Clean the data based on the findings from the data exploration step. This includes handling missing values and potentially converting data types for columns identified as needing it (e.g., object type columns that should be numerical).


**Reasoning**:
Based on the previous exploration, several columns have missing values and some object type columns need to be converted to numeric. I will address the missing values by filling them where appropriate and convert the data types. For numerical columns with missing values, I will fill them with 0, as missing financial or numerical data often implies a value of zero. For categorical/date columns, I will also fill with an empty string or a placeholder if necessary, but based on the column names, numerical imputation seems more relevant. I will then convert the relevant object columns to numeric, handling potential errors.



```python
# Fill missing numerical values with 0
numerical_cols_with_missing = ['nilai_perolehan', 'umur_ekonomis', 'nilai_residu', 'penyusutan_fiskal_tahunan', 'tax_rate', 'insentif_pajak']
for col in numerical_cols_with_missing:
    if col in df.columns:
        df[col] = df[col].fillna(0)

# Fill missing categorical/date values with an empty string or a placeholder if appropriate
categorical_cols_with_missing = ['metode_penyusutan', 'tanggal_perolehan', 'tax_holiday_awal', 'tax_holiday_akhir']
for col in categorical_cols_with_missing:
     if col in df.columns:
        df[col] = df[col].fillna('') # Using empty string for simplicity

# Identify object columns that should be numerical and convert them
# Based on the head and dtypes, 'nilai_perolehan', 'umur_ekonomis', 'nilai_residu', 'penyusutan_fiskal_tahunan', 'tax_rate', 'insentif_pajak' are already numeric (float64),
# but let's double check and handle potential non-numeric values if they were object type initially.
# Since they are already float64, the primary need was filling NaNs, which is done above.
# However, if there were object columns like 'some_numeric_col' with non-numeric entries,
# the conversion logic would be as follows:
# for col in ['some_numeric_col_if_exists']: # Replace with actual column names if any
#     if col in df.columns and df[col].dtype == 'object':
#         df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0) # Convert to numeric, coerce errors to NaN, then fill NaN with 0

# Verify the changes
print("\nMissing values after cleaning:")
print(df.isnull().sum())
print("\nData types after cleaning:")
print(df.dtypes)
```

**Reasoning**:
The previous step successfully handled the missing values by filling them. The output shows that there are no more missing values. However, the `insentif_pajak` column is still of object type but contains numerical values (0 and potentially others based on the data). It should be converted to a numerical type for analysis. The other object columns (`scenario`, `aset_id`, `kategori`, `metode_penyusutan`, `tanggal_perolehan`, `tax_holiday_awal`, `tax_holiday_akhir`) are appropriately of object type and do not require conversion for this subtask.



```python
# Convert 'insentif_pajak' to numeric, coercing errors
df['insentif_pajak'] = pd.to_numeric(df['insentif_pajak'], errors='coerce').fillna(0)

# Verify the changes
print("\nData types after converting insentif_pajak:")
print(df.dtypes)
```

## Data wrangling

### Subtask:
Prepare the data for analysis and visualization by performing necessary transformations or creating new features if required. This might involve extracting year information from date columns, or aggregating data at a higher level if the analysis plan requires it.


**Reasoning**:
Prepare the data by converting 'tahun' to numeric, ensuring other numeric columns are correct types, and extracting the year from 'tanggal_perolehan'.



```python
# Convert 'tahun' to numeric, coercing errors to NaN and filling with 0
df['tahun'] = pd.to_numeric(df['tahun'], errors='coerce').fillna(0).astype(int)

# Ensure relevant numerical columns are numeric.
# Based on previous steps, these are already float64 or int64 and NaNs were handled.
# We can just confirm their types.
numeric_cols_to_check = ['nilai_perolehan', 'penyusutan_fiskal_tahunan', 'insentif_pajak', 'tax_rate']
for col in numeric_cols_to_check:
    if col in df.columns:
        # This will raise an error if conversion is not possible after filling NaNs,
        # indicating a data issue that needs investigation if it occurs.
        # Since NaNs are filled with 0, this should be safe.
        df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0)

# Extract year from 'tanggal_perolehan'
# First, convert to datetime, coercing errors
df['tanggal_perolehan'] = pd.to_datetime(df['tanggal_perolehan'], errors='coerce')
# Extract year, NaT values will result in NaN
df['tahun_perolehan'] = df['tanggal_perolehan'].dt.year
# Fill resulting NaNs with a placeholder, e.g., 0 or -1, or a specific future year if appropriate.
# Using 0 as it aligns with potential filling of 'tahun' column.
df['tahun_perolehan'] = df['tahun_perolehan'].fillna(0).astype(int)


# Verify the changes
print("\nData types after transformation:")
print(df.dtypes)
display(df[['tahun', 'tanggal_perolehan', 'tahun_perolehan', 'nilai_perolehan', 'penyusutan_fiskal_tahunan', 'insentif_pajak', 'tax_rate']].head())
```

## Data analysis

### Subtask:
Analyze the data to gain insights. This could involve grouping data by relevant categories (e.g., year, industry), calculating aggregates (e.g., total tax revenue), or identifying trends over time.


**Reasoning**:
Group the data by year and calculate the sum of the specified columns, then display the head of the resulting DataFrame.



```python
# Calculate 'estimasi_pajak'
df['estimasi_pajak'] = df['penyusutan_fiskal_tahunan'] * df['tax_rate']

# 1. Group by 'tahun' and calculate the sum
df_yearly_summary = df.groupby('tahun').agg({
    'penyusutan_fiskal_tahunan': 'sum',
    'insentif_pajak': 'sum',
    'estimasi_pajak': 'sum'
}).reset_index()
display(df_yearly_summary.head())
```

**Reasoning**:
Group the data by both 'tahun' and 'kategori' and calculate the sum of the specified columns, then display the head of the resulting DataFrame.



```python
# 2. Group by 'tahun' and 'kategori' and calculate the sum if 'kategori' column exists
if 'kategori' in df.columns:
    df_yearly_category_summary = df.groupby(['tahun', 'kategori']).agg({
        'penyusutan_fiskal_tahunan': 'sum',
        'insentif_pajak': 'sum',
        'estimasi_pajak': 'sum'
    }).reset_index()
    display(df_yearly_category_summary.head())
else:
    df_yearly_category_summary = pd.DataFrame() # Create empty dataframe if 'kategori' doesn't exist
    print("Column 'kategori' not found in the DataFrame. Skipping yearly category summary.")
```

**Reasoning**:
Calculate the total 'estimasi_pajak' across all years and categories, and display the result.



```python
# 3. Calculate the total 'estimasi_pajak' across all years and categories
total_estimasi_pajak = df['estimasi_pajak'].sum()
print(f"\nTotal Estimasi Pajak across all years and categories: {total_estimasi_pajak:.2f}")
```

## Data visualization

### Subtask:
Create visualizations of the data to highlight the findings from the analysis. This should include colorful charts as requested, focusing on presenting the data clearly and informatively.


**Reasoning**:
Import the necessary libraries for plotting.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure plots are displayed inline
%matplotlib inline
```

**Reasoning**:
Create a bar chart to visualize the total 'estimasi_pajak' per year using the `df_yearly_summary` DataFrame, as per instruction 2.



```python
plt.figure(figsize=(10, 6))
sns.barplot(x='tahun', y='estimasi_pajak', data=df_yearly_summary, palette='viridis')
plt.title('Total Estimasi Pajak per Tahun')
plt.xlabel('Tahun')
plt.ylabel('Estimasi Pajak')
plt.grid(axis='y')
plt.show()
```

**Reasoning**:
Create a stacked bar chart to visualize 'estimasi_pajak' by 'tahun' and 'kategori' using the `df_yearly_category_summary` DataFrame, as per instruction 3.



```python
if not df_yearly_category_summary.empty:
    plt.figure(figsize=(12, 7))
    sns.histplot(data=df_yearly_category_summary, x='tahun', weights='estimasi_pajak', hue='kategori', multiple='stack', palette='viridis', shrink=0.8)
    plt.title('Estimasi Pajak by Tahun and Kategori')
    plt.xlabel('Tahun')
    plt.ylabel('Estimasi Pajak')
    plt.xticks(df_yearly_category_summary['tahun'].unique()) # Ensure all years are shown as ticks
    plt.grid(axis='y')
    plt.legend(title='Kategori') # Add legend for categories
    plt.show()
else:
    print("df_yearly_category_summary is empty. Skipping stacked bar chart.")
```

**Reasoning**:
Create a line plot to show the trend of 'penyusutan_fiskal_tahunan' and 'insentif_pajak' over the years using the `df_yearly_summary` DataFrame, as per instruction 4.



```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='tahun', y='penyusutan_fiskal_tahunan', data=df_yearly_summary, marker='o', label='Penyusutan Fiskal Tahunan')
sns.lineplot(x='tahun', y='insentif_pajak', data=df_yearly_summary, marker='o', label='Insentif Pajak')
plt.title('Tren Penyusutan Fiskal Tahunan dan Insentif Pajak per Tahun')
plt.xlabel('Tahun')
plt.ylabel('Jumlah')
plt.xticks(df_yearly_summary['tahun']) # Ensure all years are shown as ticks
plt.grid(True)
plt.legend()
plt.show()
```
