## Data cleaning

### Subtask:
Handle missing values in the DataFrame based on the exploration results.


**Reasoning**:
Fill the missing values in the specified columns using appropriate strategies and fill the tax holiday columns with a placeholder. Then, verify that all missing values are handled.



```python
# Fill missing numerical values with the median
numerical_cols_to_fill = ['nilai_perolehan', 'umur_ekonomis', 'nilai_residu', 'penyusutan_fiskal_tahunan', 'tax_rate', 'insentif_pajak']
for col in numerical_cols_to_fill:
    df[col] = df[col].fillna(df[col].median())

# Fill missing categorical/object values with the mode
categorical_cols_to_fill = ['metode_penyusutan', 'tanggal_perolehan']
for col in categorical_cols_to_fill:
    df[col] = df[col].fillna(df[col].mode()[0])

# Fill missing values in tax holiday columns with 'Not Applicable'
df['tax_holiday_awal'] = df['tax_holiday_awal'].fillna('Not Applicable')
df['tax_holiday_akhir'] = df['tax_holiday_akhir'].fillna('Not Applicable')

# Verify that all missing values have been handled
print("Missing values after handling:")
print(df.isnull().sum())
```

## Data wrangling

### Subtask:
Convert the 'tanggal_perolehan' column to datetime objects and extract the year, and convert 'tax_rate' to a percentage.


**Reasoning**:
Convert 'tanggal_perolehan' to datetime, extract the year, and convert 'tax_rate' to a percentage as instructed.



```python
df['tanggal_perolehan'] = pd.to_datetime(df['tanggal_perolehan'], errors='coerce')
df['tahun_perolehan'] = df['tanggal_perolehan'].dt.year
df['tax_rate'] = df['tax_rate'] * 100
display(df[['tanggal_perolehan', 'tahun_perolehan', 'tax_rate']].head())
```

## Data analysis

### Subtask:
Analyze the trends of 'pendapatan', 'laba_sebelum_pajak', and 'pph_terutang' over the years. Also, examine the distribution of 'tax_rate' and 'insentif_pajak'.


**Reasoning**:
Group the DataFrame by the 'tahun' column and calculate the sum of 'pendapatan', 'laba_sebelum_pajak', and 'pph_terutang' for each year, then calculate descriptive statistics for 'tax_rate' and frequency distribution for 'insentif_pajak'.



```python
yearly_trends = df.groupby('tahun')[['pendapatan', 'laba_sebelum_pajak', 'pph_terutang']].sum()
print("Yearly Trends:")
display(yearly_trends)

tax_rate_stats = df['tax_rate'].describe()
print("\nTax Rate Descriptive Statistics:")
display(tax_rate_stats)

insentif_pajak_distribution = df['insentif_pajak'].value_counts()
print("\nInsentif Pajak Distribution:")
display(insentif_pajak_distribution)
```

## Data visualization

### Subtask:
Create visualizations to represent the findings from the data analysis, specifically focusing on the yearly trends of 'pendapatan', 'laba_sebelum_pajak', and 'pph_terutang', and the distribution of 'tax_rate' and 'insentif_pajak'.


**Reasoning**:
Plot the yearly trends of 'pendapatan', 'laba_sebelum_pajak', and 'pph_terutang' using a line plot with different colors and appropriate labels.



```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(yearly_trends.index, yearly_trends['pendapatan'], marker='o', label='Pendapatan', color='blue')
plt.plot(yearly_trends.index, yearly_trends['laba_sebelum_pajak'], marker='o', label='Laba Sebelum Pajak', color='green')
plt.plot(yearly_trends.index, yearly_trends['pph_terutang'], marker='o', label='PPh Terutang', color='red')
plt.title('Tren Tahunan Pendapatan, Laba Sebelum Pajak, dan PPh Terutang')
plt.xlabel('Tahun')
plt.ylabel('Jumlah')
plt.grid(True)
plt.legend()
plt.show()
```

**Reasoning**:
Create a histogram to visualize the distribution of the 'tax_rate' with appropriate labels and a title.



```python
plt.figure(figsize=(8, 6))
plt.hist(df['tax_rate'], bins=10, color='purple', edgecolor='black')
plt.title('Distribusi Tax Rate')
plt.xlabel('Tax Rate (%)')
plt.ylabel('Frekuensi')
plt.grid(True)
plt.show()
```

**Reasoning**:
Generate a bar plot to show the distribution of 'insentif_pajak' with appropriate labels and a title.



```python
plt.figure(figsize=(10, 6))
insentif_pajak_distribution.plot(kind='bar', color='orange')
plt.title('Distribusi Insentif Pajak')
plt.xlabel('Jenis Insentif Pajak')
plt.ylabel('Jumlah')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
